```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.arange(10)
Y=(X-5)**2
print("X : ",X,"Y :",Y)
```

```python
plt.plot(X,Y)
plt.ylabel=("X")
plt.xlabel=("F(x)")
plt.show()
```

```python
x = 0
lr = 0.1
error=[]
plt.plot(X,Y)
for i in range(50):
    grad=2*(x-5) #dif of Y is gradient
    x=x-lr*grad
    y=(x-5)**2
    error.append(y)
    plt.scatter(x,y)
    print(x)
   
```

```python
plt.plot(error)
plt.show()
```
